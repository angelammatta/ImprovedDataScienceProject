```python
import pandas as pd
import matplotlib.pyplot as plt


#urls for Github data

urlAVGSubFin = 'https://raw.githubusercontent.com/angelammatta/ImprovedDataScienceProject/main/Average%20Subsequent%20Private%20Financing.csv'
urlAVGOrigSSBCI = 'https://raw.githubusercontent.com/angelammatta/ImprovedDataScienceProject/main/Average%20SSBCI%20Original%20Funds.csv'
urlAVGRevenue = 'https://raw.githubusercontent.com/angelammatta/ImprovedDataScienceProject/main/Average%20Revenue.csv'
urlAVGLoanInvest = 'https://raw.githubusercontent.com/angelammatta/ImprovedDataScienceProject/main/Average%20Loan%20Investment%20Amount.csv'
urlAVGConFin = 'https://raw.githubusercontent.com/angelammatta/ImprovedDataScienceProject/main/Average%20Concurrent%20Private%20Financing.csv'



# read the csv file to extract data


#AVG Sub Fin 
dataAVGSubFin = pd.read_csv(urlAVGSubFin)
stateName = dataAVGSubFin['state_id']
AVGSubFin = dataAVGSubFin['Average Subsequent Private Financing']

fig = plt.figure(figsize=(50, 20))

plt.scatter(stateName, AVGSubFin)

plt.rcParams.update({'font.size': 25})

plt.xticks(rotation=40)

plt.title('Average Subsequent Private Financing')
plt.xlabel('State Name')
plt.ylabel('Average Subsequent Private Financing(in hundreds of thousands)')


plt.show()

##


#AVG Orig SSBCI
dataAVGSubFin = pd.read_csv(urlAVGOrigSSBCI)
stateName = dataAVGSubFin['state_id']
AVGSubFin = dataAVGSubFin['Average SSBCI Original Funds']

fig = plt.figure(figsize=(50, 20))

plt.scatter(stateName, AVGSubFin)

plt.rcParams.update({'font.size': 25})

plt.xticks(rotation=40)

plt.title('Average SSBCI Original Funds')
plt.xlabel('State Name')
plt.ylabel('Average SSBCI Original Funds(in hundreds of thousands)')


plt.show()


#AVG revenue
dataAVGRevenue = pd.read_csv(urlAVGRevenue)
stateName = dataAVGRevenue['state_id']
AVGRevenue = dataAVGRevenue['Average Revenue']

fig = plt.figure(figsize=(50, 20))

plt.scatter(stateName, AVGRevenue)

plt.rcParams.update({'font.size': 25})

plt.xticks(rotation=40)

plt.title('Average Revenue')
plt.xlabel('State Name')
plt.ylabel('Average Revenue(in hundreds of thousands)')


plt.show()



#AVG Loan Invest
dataAVGLoanInvest = pd.read_csv(urlAVGLoanInvest)
stateName = dataAVGLoanInvest['state_id']
AVGLoanInvest = dataAVGLoanInvest['Average Loan Investment Amount']

fig = plt.figure(figsize=(50, 20))

plt.scatter(stateName, AVGLoanInvest)

plt.rcParams.update({'font.size': 25})

plt.xticks(rotation=40)

plt.title('Average Loan Investment')
plt.xlabel('State Name')
plt.ylabel('Average Loan Investment Amount(in hundreds of thousands)')


plt.show()


#AVG Concurrent financing
dataAVGConFin = pd.read_csv(urlAVGConFin)
stateName = dataAVGConFin['state_id']
AVGConFin = dataAVGConFin['Average Concurrent Private Financing']

fig = plt.figure(figsize=(50, 20))

plt.scatter(stateName, AVGConFin)

plt.rcParams.update({'font.size': 25})

plt.xticks(rotation=40)

plt.title('Average Concurrent Private Financing')
plt.xlabel('State Name')
plt.ylabel('Average Concurrent Private Financing(in hundreds of thousands)')


plt.show()

```
